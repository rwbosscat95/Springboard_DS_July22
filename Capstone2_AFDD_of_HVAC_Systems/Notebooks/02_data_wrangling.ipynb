## 2.4.1 mzvav_1_data <a id='2.4.1_mzvav_1_data'></a>

```python
mzvav_1_data['Datetime'] = pd.to_datetime(mzvav_1_data['Datetime'])
```

```python
# Define the unfault data of mzvav_1_data - mzvav_1_unfaulted with 4 weeks
mzvav_1_unfaulted1 = mzvav_1_data.loc[(mzvav_1_data['Datetime'] >= datetime(2017,1,30,0,0,0)) & 
                                     (mzvav_1_data['Datetime'] < datetime(2017,2,6,0,0,0))]
mzvav_1_unfaulted2 = mzvav_1_data.loc[(mzvav_1_data['Datetime'] >= datetime(2017,5,2,0,0,0)) & 
                                     (mzvav_1_data['Datetime'] < datetime(2017,5,8,0,0,0))]
mzvav_1_unfaulted3 = mzvav_1_data.loc[(mzvav_1_data['Datetime'] >= datetime(2017,7,31,0,0,0)) & 
                                     (mzvav_1_data['Datetime'] < datetime(2017,8,7,0,0,0))]
mzvav_1_unfaulted4 = mzvav_1_data.loc[(mzvav_1_data['Datetime'] >= datetime(2017,10,30,0,0,0)) & 
                                     (mzvav_1_data['Datetime'] < datetime(2017,11,6,0,0,0))]
mzvav_1_unfaulted = pd.concat([mzvav_1_unfaulted1, mzvav_1_unfaulted2, mzvav_1_unfaulted3, mzvav_1_unfaulted4], axis=0)
```

```python
# mzvav_1_unfaulted data from mzvav_1_data with 4 weeks
print(mzvav_1_unfaulted.info())
```

```python
# Define the first fault of mzvav_1_data - mzvav_1_oaplus2 (OA temp bias +2 celcs degrees) with 4 weeks
mzvav_1_oaplus2_1 = mzvav_1_data.loc[(mzvav_1_data['Datetime'] >= datetime(2017,2,13,0,0,0)) & 
                                     (mzvav_1_data['Datetime'] < datetime(2017,2,20,0,0,0))]
mzvav_1_oaplus2_2 = mzvav_1_data.loc[(mzvav_1_data['Datetime'] >= datetime(2017,5,16,0,0,0)) & 
                                     (mzvav_1_data['Datetime'] < datetime(2017,5,22,0,0,0))]
mzvav_1_oaplus2_3 = mzvav_1_data.loc[(mzvav_1_data['Datetime'] >= datetime(2017,8,14,0,0,0)) & 
                                     (mzvav_1_data['Datetime'] < datetime(2017,8,21,0,0,0))]
mzvav_1_oaplus2_4 = mzvav_1_data.loc[(mzvav_1_data['Datetime'] >= datetime(2017,11,13,0,0,0)) & 
                                     (mzvav_1_data['Datetime'] < datetime(2017,11,20,0,0,0))]
mzvav_1_oaplus2 = pd.concat([mzvav_1_oaplus2_1, mzvav_1_oaplus2_2, mzvav_1_oaplus2_3, mzvav_1_oaplus2_4], axis=0)
```

```python
print(mzvav_1_oaplus2.info())
```

```python
# Combined the mzvav_1_oaplus2 with mzvav_1_unfaulted
mzvav_1_oaplus2_unfaulted = pd.concat([mzvav_1_oaplus2, mzvav_1_unfaulted], axis=0)
print(mzvav_1_oaplus2_unfaulted.info())
```
